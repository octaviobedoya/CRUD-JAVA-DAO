# ANALISIS RFM (Recency,Frequency,Monetary)

```python
import pandas as pd
import matplotlib.pyplot as plt

## Versión simplificada y robusta del análisis RFM
def analisis_rfm_reducido(df):
    # 1. Preparación básica de datos
    df['TotalValue'] = df['Quantity'] * df['UnitPrice']
    analysis_date = df['InvoiceDate'].max() + pd.Timedelta(days=1)

    # 2. Cálculo RFM básico con manejo de errores
    rfm = df.groupby('CustomerID').agg({
        'InvoiceDate': lambda x: (analysis_date - x.max()).days,
        'InvoiceNo': 'nunique',
        'TotalValue': 'sum'
    }).rename(columns={
        'InvoiceDate': 'Recency',
        'InvoiceNo': 'Frequency',
        'TotalValue': 'Monetary'
    }).reset_index()

    # 3. Segmentación simplificada sin percentiles problemáticos
    # Usamos cortes directos en lugar de qcut para evitar errores
    rfm['Segment'] = 'Regular'
    rfm.loc[rfm['Recency'] <= 30, 'Segment'] = 'Reciente'
    rfm.loc[rfm['Frequency'] > rfm['Frequency'].median(), 'Segment'] = 'Frecuente'
    rfm.loc[rfm['Monetary'] > rfm['Monetary'].quantile(0.75), 'Segment'] = 'Valioso'

    # Combinaciones
    rfm.loc[(rfm['Recency'] <= 30) &
           (rfm['Frequency'] > rfm['Frequency'].median()) &
           (rfm['Monetary'] > rfm['Monetary'].quantile(0.75)), 'Segment'] = 'Mejor Cliente'

    rfm.loc[rfm['Recency'] > 365, 'Segment'] = 'Inactivo'

    # 4. Informe visual mínimo
    plt.figure(figsize=(10, 5))
    rfm['Segment'].value_counts().plot(kind='bar', color='skyblue')
    plt.title('Distribución de Segmentos RFM')
    plt.ylabel('Número de Clientes')
    plt.xticks(rotation=45)
    plt.show()

    # 5. Estadísticas clave
    print("\n📊 RESUMEN RFM")
    print("="*30)
    print(f"Clientes totales: {len(rfm)}")
    print(f"Recencia promedio: {rfm['Recency'].mean():.1f} días")
    print(f"Frecuencia promedio: {rfm['Frequency'].mean():.1f} compras")
    print(f"Valor promedio: ${rfm['Monetary'].mean():,.2f}")

    print("\n🔝 Top 3 clientes por valor:")
    print(rfm.sort_values('Monetary', ascending=False).head(3)[['CustomerID', 'Segment', 'Monetary']])

    return rfm

# Ejecutar análisis
rfm_result = analisis_rfm_reducido(minorista_df)
```

Identificar problemas específicos en el dataset Online Retail con visualizaciones y métricas detalladas.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Configuración inicial corregida (sin dependencia de estilos)
plt.style.use('default')  # Usamos el estilo por defecto para evitar errores
sns.set_style("whitegrid")  # Estilo alternativo que sí está disponible

# Función para análisis de valores nulos
def analizar_nulos(df):
    print("🔍 ANÁLISIS DE VALORES NULOS")
    print("=" * 50)

    nulos = df.isnull().sum().sort_values(ascending=False)
    porcentaje_nulos = (nulos / len(df)) * 100
    resumen_nulos = pd.DataFrame({"Valores Nulos": nulos, "% del Total": porcentaje_nulos.round(2)})

    plt.figure(figsize=(10, 4))
    sns.barplot(x=porcentaje_nulos.values, y=porcentaje_nulos.index)
    plt.title("Porcentaje de Valores Nulos por Columna")
    plt.xlabel("% de Nulos")
    plt.show()

    print(resumen_nulos[resumen_nulos["Valores Nulos"] > 0])
    print("\n")

# Función para análisis de anomalías numéricas
def analizar_anomalias(df):
    print("📊 ANÁLISIS DE ANOMALÍAS NUMÉRICAS")
    print("=" * 50)

    print("\n📦 Estadísticas de Quantity:")
    print(df['Quantity'].describe())

    print("\n💰 Estadísticas de UnitPrice:")
    print(df['UnitPrice'].describe())

    fig, ax = plt.subplots(1, 2, figsize=(12, 4))
    sns.boxplot(data=df, y='Quantity', ax=ax[0])
    ax[0].set_title("Distribución de Quantity")
    sns.boxplot(data=df, y='UnitPrice', ax=ax[1])
    ax[1].set_title("Distribución de UnitPrice")
    plt.tight_layout()
    plt.show()

    # Detección de outliers con método IQR
    def calcular_outliers(serie):
        q1, q3 = serie.quantile([0.25, 0.75])
        iqr = q3 - q1
        return serie[(serie < (q1 - 1.5 * iqr)) | (serie > (q3 + 1.5 * iqr))]

    outliers_q = calcular_outliers(df['Quantity'])
    outliers_p = calcular_outliers(df['UnitPrice'])

    print(f"\n🚨 Outliers en Quantity: {len(outliers_q)} ({(len(outliers_q)/len(df)*100):.2f}%)")
    print(f"🚨 Outliers en UnitPrice: {len(outliers_p)} ({(len(outliers_p)/len(df)*100):.2f}%)")

    if not outliers_q.empty:
        print("\nEjemplos de outliers en Quantity:")
        print(df.loc[outliers_q.index, ['InvoiceNo', 'StockCode', 'Description', 'Quantity']].head(3))

    if not outliers_p.empty:
        print("\nEjemplos de outliers en UnitPrice:")
        print(df.loc[outliers_p.index, ['StockCode', 'Description', 'UnitPrice']].head(3))

# Función para análisis de devoluciones
def analizar_devoluciones(df):
    print("🔄 ANÁLISIS DE DEVOLUCIONES Y CANCELACIONES")
    print("=" * 50)

    devoluciones = df[df['Quantity'] < 0]
    print(f"📉 Devoluciones detectadas: {len(devoluciones)} ({(len(devoluciones)/len(df)*100):.2f}%)")

    facturas_canceladas = df[df['InvoiceNo'].astype(str).str.contains('^[CA]', regex=True)]
    print(f"❌ Facturas canceladas: {len(facturas_canceladas)} ({(len(facturas_canceladas)/len(df)*100):.2f}%)")

    if not devoluciones.empty:
        print("\n🔍 Top 5 productos con más devoluciones:")
        print(devoluciones.groupby('Description')['Quantity'].sum().sort_values().head(5))

    if not facturas_canceladas.empty:
        print("\n🔍 Ejemplo de facturas canceladas:")
        print(facturas_canceladas[['InvoiceNo', 'CustomerID', 'Quantity', 'UnitPrice']].head(3))

# Función para análisis de fechas
def analizar_fechas(df):
    print("📅 ANÁLISIS DE CONSISTENCIA EN FECHAS")
    print("=" * 50)

    if not pd.api.types.is_datetime64_any_dtype(df['InvoiceDate']):
        df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'], errors='coerce')

    fecha_min = df['InvoiceDate'].min()
    fecha_max = df['InvoiceDate'].max()
    print(f"Rango de fechas: {fecha_min} a {fecha_max}")

    fechas_nulas = df[df['InvoiceDate'].isnull()]
    print(f"\n📌 Fechas nulas o inválidas: {len(fechas_nulas)}")

# Función principal de diagnóstico
def diagnostico_completo(df):
    print("🛠️ DIAGNÓSTICO AVANZADO DE CALIDAD DE DATOS")
    print("=" * 60)

    try:
        analizar_nulos(df)
        analizar_anomalias(df)
        analizar_devoluciones(df)
        analizar_fechas(df)
        print("\n✅ Diagnóstico completado exitosamente.")
    except Exception as e:
        print(f"\n❌ Error durante el diagnóstico: {str(e)}")

# Verificar si el DataFrame existe antes de ejecutar
try:
    if 'minorista_df' in globals():
        diagnostico_completo(minorista_df)
    else:
        print("❌ Error: No se encontró el DataFrame 'minorista_df'.")
        print("Por favor carga tus datos primero con:")
        print("""
# Ejemplo de carga de datos:
import pandas as pd
from google.colab import files
import io

uploaded = files.upload()
file_name = next(iter(uploaded))
minorista_df = pd.read_csv(io.BytesIO(uploaded[file_name]))
""")
except NameError:
    print("❌ Error: El DataFrame 'minorista_df' no está definido.")
```

```python
import pandas as pd
from google.colab import files
import io

def cargar_dataset_retail():
    print("📤 Subiendo archivo...")
    uploaded = files.upload()
    file_name = next(iter(uploaded))

    try:
        # Primero leer sin parsear fechas para inspeccionar columnas
        df_temp = pd.read_csv(
            io.BytesIO(uploaded[file_name]),
            encoding='ISO-8859-1',
            sep=',',
            nrows=5  # Solo leer primeras filas para inspección
        )

        print("\n🔍 Columnas detectadas en el archivo:")
        print(df_temp.columns.tolist())

        # Verificar nombre correcto de columna de fecha
        date_col = None
        for col in df_temp.columns:
            if 'date' in col.lower() or 'invoice' in col.lower():
                date_col = col
                break

        if date_col is None:
            print("\n⚠️ No se encontró columna de fecha. Usando primera columna.")
            date_col = df_temp.columns[0]

        print(f"\n🔧 Se usará '{date_col}' como columna de fecha")

        # Leer el archivo completo con los parámetros correctos
        df = pd.read_csv(
            io.BytesIO(uploaded[file_name]),
            encoding='ISO-8859-1',
            sep=',',
            quotechar='"',
            dtype={'CustomerID': str},
            parse_dates=[date_col],  # Usar la columna detectada
            on_bad_lines='warn'
        )

        # Renombrar columnas para consistencia
        df.columns = df.columns.str.strip()
        if date_col != 'InvoiceDate':
            df = df.rename(columns={date_col: 'InvoiceDate'})

        print(f"\n✅ Dataset cargado correctamente con {len(df):,} registros")
        print("\n📋 Estructura del dataset:")
        print(df.info())

        return df

    except Exception as e:
        print(f"\n❌ Error al cargar el archivo: {str(e)}")
        print("\n💡 Posibles soluciones:")
        print("1. Verifica que el archivo sea el correcto")
        print("2. Revisa los delimitadores (puede usar ';' en lugar de ',')")
        print("3. Prueba con encoding diferente como 'latin1' o 'utf-8'")
        return None

# Cargar el dataset
minorista_df = cargar_dataset_retail()

if minorista_df is not None:
    print("\n🔍 Primeras filas del dataset:")
    display(minorista_df.head(3))
```

# LIMPIEZA Y TRANSFORMACION AVANZADA DE DATOS
Limpieza:

Eliminación de duplicados exactos.

Imputación de nulos (para CustomerID, Description).

Filtrado de outliers estadísticos.

Transformación:

Creación de variables derivadas (TotalValue = Quantity * UnitPrice).

Discretización (ej: categorizar clientes por frecuencia de compra).

Codificación de texto (ej: países a códigos categóricos).

```python
import pandas as pd
import numpy as np
from datetime import datetime

def limpiar_transformar(df):
    # --- 1. Verificar columnas existentes ---
    print("🔍 Verificando estructura del dataset...")
    columnas_requeridas = {'InvoiceNo', 'StockCode', 'Description', 'Quantity',
                          'InvoiceDate', 'UnitPrice', 'CustomerID', 'Country'}
    columnas_faltantes = columnas_requeridas - set(df.columns)

    if columnas_faltantes:
        print(f"⚠️ Advertencia: Columnas faltantes: {columnas_faltantes}")

    # --- 2. Copia segura ---
    df_clean = df.copy()

    # --- 3. Manejo de valores nulos (con verificación) ---
    print("\n🔄 Manejo de valores nulos...")
    if 'CustomerID' in df_clean.columns:
        print(f"- Registros antes de eliminar CustomerID nulos: {len(df_clean):,}")
        df_clean = df_clean[df_clean['CustomerID'].notna()]
        print(f"- Registros después: {len(df_clean):,}")
    else:
        print("- No se encontró columna CustomerID")

    if 'Description' in df_clean.columns:
        df_clean['Description'] = df_clean['Description'].fillna('UNKNOWN').str.strip()

    # --- 4. Limpieza de texto ---
    print("\n✂️ Normalizando texto...")
    text_cols = ['Description', 'Country'] if 'Country' in df_clean.columns else ['Description']
    for col in text_cols:
        if col in df_clean.columns:
            df_clean[col] = df_clean[col].astype(str).str.upper().str.strip()

    # --- 5. Manejo de cantidades y precios ---
    print("\n🧮 Limpiando valores numéricos...")
    if 'Quantity' in df_clean.columns:
        # Separar devoluciones
        df_devoluciones = df_clean[df_clean['Quantity'] < 0].copy()
        df_clean = df_clean[df_clean['Quantity'] >= 0]

        # Limitar outliers (percentil 99)
        q_upper = df_clean['Quantity'].quantile(0.99)
        df_clean['Quantity'] = df_clean['Quantity'].clip(upper=q_upper)
    else:
        df_devoluciones = pd.DataFrame()
        print("- No se encontró columna Quantity")

    if 'UnitPrice' in df_clean.columns:
        # Eliminar precios cero o negativos
        df_clean = df_clean[df_clean['UnitPrice'] > 0]

        # Limitar outliers
        p_upper = df_clean['UnitPrice'].quantile(0.99)
        df_clean['UnitPrice'] = df_clean['UnitPrice'].clip(upper=p_upper)

    # --- 6. Transformación de fechas ---
    print("\n📅 Estandarizando fechas...")
    if 'InvoiceDate' in df_clean.columns:
        df_clean['InvoiceDate'] = pd.to_datetime(df_clean['InvoiceDate'], errors='coerce')
        df_clean = df_clean[df_clean['InvoiceDate'].notna()]
    else:
        print("- No se encontró columna InvoiceDate")

    # --- 7. Variables derivadas ---
    print("\n🧰 Creando variables derivadas...")
    if all(col in df_clean.columns for col in ['Quantity', 'UnitPrice']):
        df_clean['TotalValue'] = df_clean['Quantity'] * df_clean['UnitPrice']

    if 'InvoiceDate' in df_clean.columns:
        df_clean['YearMonth'] = df_clean['InvoiceDate'].dt.to_period('M')

    # --- 8. Resultados ---
    print("\n✅ Limpieza completada:")
    print(f"- Registros originales: {len(df):,}")
    print(f"- Registros limpios: {len(df_clean):,}")
    print(f"- Devoluciones identificadas: {len(df_devoluciones):,}")

    return df_clean, df_devoluciones

# Ejecución segura
if 'minorista_df' in globals():
    try:
        df_limpio, devoluciones = limpiar_transformar(minorista_df)

        print("\n🔍 Muestra de datos limpios:")
        display(df_limpio.head(3))

        print("\n📊 Resumen post-limpieza:")
        print(df_limpio.info())
    except Exception as e:
        print(f"❌ Error durante la limpieza: {str(e)}")
else:
    print("❌ Error: No se encontró el DataFrame 'minorista_df'")
    print("Por favor carga tus datos primero con:")
    print("""
    from google.colab import files
    uploaded = files.upload()
    file_name = next(iter(uploaded))
    minorista_df = pd.read_csv(io.BytesIO(uploaded[file_name]))
    """)
```

```python
import pandas as pd
from google.colab import files

def exportar_archivos_limpios(df_limpio, devoluciones):
    print("📤 Preparando archivos para descarga...")

    try:
        # 1. Exportar dataset limpio principal
        nombre_principal = 'online_retail_clean.csv'
        df_limpio.to_csv(nombre_principal, index=False, encoding='utf-8')
        print(f"✅ Dataset limpio guardado como: {nombre_principal}")

        # 2. Exportar devoluciones (si existen)
        if len(devoluciones) > 0:
            nombre_devoluciones = 'online_retail_returns.csv'
            devoluciones.to_csv(nombre_devoluciones, index=False, encoding='utf-8')
            print(f"✅ Devoluciones guardadas como: {nombre_devoluciones}")

        # 3. Crear archivo ZIP con ambos
        import zipfile
        nombre_zip = 'online_retail_clean_data.zip'
        with zipfile.ZipFile(nombre_zip, 'w') as zipf:
            zipf.write(nombre_principal)
            if len(devoluciones) > 0:
                zipf.write(nombre_devoluciones)
        print(f"📦 Paquete completo guardado como: {nombre_zip}")

        # 4. Opciones de descarga
        print("\n💾 Elige cómo descargar:")
        print("1. Descargar solo dataset limpio:")
        print(f"   files.download('{nombre_principal}')")

        if len(devoluciones) > 0:
            print("\n2. Descargar solo devoluciones:")
            print(f"   files.download('{nombre_devoluciones}')")

        print("\n3. Descargar todo (recomendado):")
        print(f"   files.download('{nombre_zip}')")

        return nombre_principal, nombre_devoluciones if len(devoluciones) > 0 else None, nombre_zip

    except Exception as e:
        print(f"❌ Error al exportar: {str(e)}")
        return None, None, None

# Ejecutar exportación (asumiendo que df_limpio y devoluciones existen)
if all(var in globals() for var in ['df_limpio', 'devoluciones']):
    print(f"🔢 Estadísticas de exportación:")
    print(f"- Registros limpios: {len(df_limpio):,}")
    print(f"- Registros de devolución: {len(devoluciones):,}")

    # Llamar a la función de exportación
    archivo_principal, archivo_devoluciones, archivo_zip = exportar_archivos_limpios(df_limpio, devoluciones)

    # Descargar automáticamente el ZIP (opcional)
    if archivo_zip:
        print("\n⏳ Descargando archivo ZIP...")
        files.download(archivo_zip)
else:
    print("❌ Error: No se encontraron los DataFrames limpios")
    print("Ejecuta primero el proceso de limpieza con limpiar_transformar()")
```

```python
import pandas as pd
from google.colab import files
import io
import zipfile

def verificar_y_exportar():
    # Verificar si los DataFrames limpios existen y tienen datos
    if 'df_limpio' not in globals() or len(df_limpio) == 0:
        print("❌ Error: No se encontró el DataFrame limpio o está vacío")
        print("Ejecuta primero el proceso de limpieza completa:")
        print("""
        df_limpio, devoluciones = limpiar_transformar(minorista_df)
        """)
        return

    # Crear archivos temporales
    try:
        print("🔄 Creando archivos de exportación...")

        # 1. Dataset principal limpio
        output_principal = io.StringIO()
        df_limpio.to_csv(output_principal, index=False)
        principal_content = output_principal.getvalue()

        # 2. Archivo de devoluciones (si existe)
        if 'devoluciones' in globals() and len(devoluciones) > 0:
            output_devoluciones = io.StringIO()
            devoluciones.to_csv(output_devoluciones, index=False)
            devoluciones_content = output_devoluciones.getvalue()

        # 3. Crear ZIP en memoria
        zip_buffer = io.BytesIO()
        with zipfile.ZipFile(zip_buffer, 'a', zipfile.ZIP_DEFLATED, False) as zip_file:
            zip_file.writestr('online_retail_clean.csv', principal_content)
            if 'devoluciones' in globals() and len(devoluciones) > 0:
                zip_file.writestr('online_retail_returns.csv', devoluciones_content)

        # Guardar ZIP
        with open('online_retail_clean.zip', 'wb') as f:
            f.write(zip_buffer.getvalue())

        print("✅ Archivos creados correctamente")
        print(f"📦 Tamaño del archivo ZIP: {len(zip_buffer.getvalue()):,} bytes")

        # Descargar automáticamente
        print("\n⏳ Descargando archivo...")
        files.download('online_retail_clean.zip')

    except Exception as e:
        print(f"❌ Error durante la exportación: {str(e)}")
        print("\n💡 Solución alternativa:")
        print("1. Ejecuta esta celda para ver los datos limpios:")
        print("display(df_limpio.head())")
        print("2. Si ves datos, prueba exportar manualmente:")
        print("df_limpio.to_csv('online_retail_clean.csv', index=False)")
        print("files.download('online_retail_clean.csv')")

# Verificar datos antes de exportar
if 'df_limpio' in globals():
    print("🔍 Verificando datos limpios...")
    print(f"Registros en df_limpio: {len(df_limpio):,}")
    if len(df_limpio) > 0:
        verificar_y_exportar()
    else:
        print("⚠️ El DataFrame limpio está vacío")
        print("Revisa el proceso de limpieza anterior")
else:
    print("❌ No se encontró df_limpio")
    print("Ejecuta primero el proceso de limpieza completo")

# Opcional: Mostrar previsualización si los datos existen
if 'df_limpio' in globals() and len(df_limpio) > 0:
    print("\n👀 Previsualización de datos limpios (primeras 3 filas):")
    display(df_limpio.head(3))
```
